```python
import pandas as pd
import numpy as np
import bar_chart_race as bcr
```

```python
df = pd.read_csv('dataset.csv')
```

```python
make = list(df['Make'])
year = list(df['Model Year'])
makes = list(df['Make'].unique())
years = list(df['Model Year'].unique())

yofm = []
temp1 = []

for j in makes:
    # taking out the total years of each make
    for i in range(len(make)):
        if j == make[i]:
            temp1.append(year[i])

    yofm.append(temp1)
    temp1 = []
print(len(years))
```

```python
#taking out the frequency of years of each make
y_count = []
temp3 = yofm[0]
temp2 = []
for i in yofm:
    for j in years:
        temp2.append(i.count(j))
    y_count.append(temp2)
    temp2 = []
len(y_count[0])
```

```python
#interchanging rows and columns 
np1 = np.array(y_count)
np2 = np1.T
np3 = np2.tolist()
len(np3[0])
```

```python
df1 = pd.DataFrame(np3, columns = makes)
df1.head()
```

```python
df1['Year'] = years
df1.set_index('Year',inplace = True)
df1.head()
```

```python
bcr.bar_chart_race(df1)
```
